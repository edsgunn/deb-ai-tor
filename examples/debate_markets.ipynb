```python
import json

from deb_ai_tor.environments import DebateMarket
```

```python
agents = [f"Agent{i}" for i in range(3)]

market = DebateMarket(
    agents=agents
)

market.reset(
    question="What is reason?",
)

actions = {
    agent: json.dumps([
        {
            "type": "buy_new_answer",
            "answer_text": f"{agent}'s answer",
            "qty": 1
        }
    ]) for agent in agents
}

observations, *_ = market.step(actions)
print(json.dumps(observations, indent=4))
```

```python
actions = {
    agent: json.dumps([
        {
            "type": "buy_answer",
            "answer_id": 0,
            "qty": 1
        }
    ]) for agent in agents
}

observations, *_ = market.step(actions)
print(json.dumps(observations, indent=4))
```
