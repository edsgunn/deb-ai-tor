```python
import json

from deb_ai_tor.environments import DebateMarket
```

```python
agents = [f"Agent{i}" for i in range(3)]

market = DebateMarket(
    agents=agents
)

market.reset(
    question="What is reason?",
)

actions = {
    agent: json.dumps([
        {
            "type": "buy_new_answer",
            "answer_text": f"{agent}'s answer",
            "amount": 0.5
        }
    ]) for agent in agents
}

observations, *_ = market.step(actions)
print(json.dumps(observations, indent=4))
```

```python
import random

actions = {}
for agent in agents:
    action_type = random.choice(["buy_answer", "sell_answer"])
    # Use agent's current balance from observations
    balance = float(observations[agent]['balance'])
    # Choose a random amount between 0.1 and the agent's balance (if selling, use holdings if available)
    if action_type == "buy_answer":
        amount = round(random.uniform(0.1, balance), 2) if balance >= 0.1 else balance
        answers = observations[agent].get('answers', [])
        answer_id = random.choice(answers)['id'] if answers else None
        actions[agent] = json.dumps([
        {
            "type": action_type,
            "answer_id": answer_id,
            "amount": amount
        }
    ])
    else:
        # For selling, assume holdings is a list of dicts with 'amount' and 'answer_id'
        holdings = observations[agent].get('holdings', [])
        if holdings:
            answer_id = random.choice(list(holdings.keys()))
            qty = round(random.uniform(0.1, holdings[answer_id]), 2) if holdings[answer_id] >= 0.1 else holdings[answer_id]
        else:
            qty = 0
            answer_id = None
        actions[agent] = json.dumps([
        {
            "type": action_type,
            "answer_id": answer_id,
            "qty": qty
        }
    ])



observations, *_ = market.step(actions)
print(json.dumps(observations, indent=4))
```
